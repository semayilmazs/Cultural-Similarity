```python
import pandas as pd
from pytrends.request import TrendReq


# Google bağlantısı
pytrends = TrendReq(hl='en-US', tz=360)

def find_correct_topic_id(keyword):
    print(f"--- '{keyword}' için Topic Arama Sonuçları ---")
    suggestions = pytrends.suggestions(keyword)
    
    if not suggestions:
        print(f"UYARI: '{keyword}' için hiçbir topic bulunamadı!")
        return

    # Bulunan tüm seçenekleri listele
    for i, s in enumerate(suggestions):
        print(f"[{i}] Başlık: {s['title']}")
        print(f"    Tip (Type): {s['type']}")
        print(f"    ID (mid):   {s['mid']}")
        print("-" * 20)

# Test etmek istediğin şüpheli veya kritik kelimeleri buraya yaz
test_keywords = ["economic crisis", "Apple"]

for kw in test_keywords:
    find_correct_topic_id(kw)
    print("\n" + "="*40 + "\n")
```
